```python
def cumsum(counts_gsmhm):
    arr = counts_gsmhm[0]/np.max(counts_gsmhm[0])
    return np.cumsum(arr)/np.max(np.cumsum(arr))

def pdf(counts_gsmhm):
    arr = counts_gsmhm[0]/np.max(counts_gsmhm[0])
    return arr
```

```python
param_arr = [agn_halos['galaxy_SMHMR_mass'], agn_halos['redshift_R'], diff_t_mm_arr[unique_halo_indicies], agn_halos['HALO_Xoff']/agn_halos['HALO_Rvir']]
string = [r'$\log{M^*}$', 'redshift', r'$T_{\rm MM}$', r'$\tilde{X}_{\rm off}$']
counts_arr = [counts_gsmhm, counts_r, counts_tmm, counts_xoff]
counts_arr_all = [counts_gsmhm_all, counts_r_all, counts_tmm_all, counts_xoff_all]

fig, ax = plt.subplots(2,2,figsize=(14,12))

ax[0,0].legend(legend_ls1, ['Selected halos', 'All halos'], bbox_to_anchor=(0.0, 1.2),  loc='upper left', fontsize=18, frameon=False, ncol=2)
ax[0,1].legend(legend_ls0, ['PDF', 'CDF'], bbox_to_anchor=(0.0, 1.2),  loc='upper left', fontsize=18, frameon=False, ncol=2)

for i, arr in enumerate(param_arr):
    #pt.plotPdf(ax[pt.axId(i)], arr, string[i], pal[i])
    ax[pt.axId(i)].plot(counts_arr[i][1][1:], pdf(counts_arr[i]), ls='-', color=pal1[i], lw=2)
    ax[pt.axId(i)].plot(counts_arr[i][1][1:], cumsum(counts_arr[i]), ls='--', color=pal1[i], lw=2)
    
    ax[pt.axId(i)].plot(counts_arr_all[i][1][1:], pdf(counts_arr_all[i]), ls='-', color=pal[i], lw=2)
    ax[pt.axId(i)].plot(counts_arr_all[i][1][1:], cumsum(counts_arr_all[i]), ls='--', color=pal[i], lw=2)

pt.labelMZTmmXoff(ax, 'Probability functions')
```

```python
def gaussian(x, amp, cen, wid):
    return (amp/ np.sqrt(2*np.pi*wid ) )* np.exp(- (x-cen)**2/ (2*wid**2))

def multipleGaussians(x, param0, param1):
    gaussian(x, param0) + gaussian(x, param1)
    
def measure(n):
    m1 = np.random.normal(size=n)
    m2 = np.random.normal(scale=0.5, size=n)
    return m1 + m2, m1 - m2
```
